```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
accelerometer = pd.read_csv('Accelerometer.csv')
accelerometer_linear = pd.read_csv('Linear Accelerometer.csv')
barometer = pd.read_csv('Barometer.csv')
gyroscope = pd.read_csv('Gyroscope.csv')
location = pd.read_csv('Location.csv')
magnetometer = pd.read_csv('Magnetometer.csv')
proximity = pd.read_csv('Proximity.csv')

print(accelerometer.shape)
print(accelerometer_linear.shape)
print(barometer.shape)
print(gyroscope.shape)
print(location.shape)
print(magnetometer.shape)
print(proximity.shape)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(accelerometer['Time (s)'], accelerometer['X (m/s^2)'], color='r', label='X')
plt.plot(accelerometer['Time (s)'], accelerometer['Y (m/s^2)'], color='g', label='Y')
plt.plot(accelerometer['Time (s)'], accelerometer['Z (m/s^2)'], color='b', label='Z')
plt.title('Accelerometer')
plt.legend(loc='upper right')
plt.show();
```

```python
plt.figure(figsize=(12, 6))
plt.plot(accelerometer_linear['Time (s)'], accelerometer_linear['X (m/s^2)'], color='r', label='X')
plt.plot(accelerometer_linear['Time (s)'], accelerometer_linear['Y (m/s^2)'], color='g', label='Y')
plt.plot(accelerometer_linear['Time (s)'], accelerometer_linear['Z (m/s^2)'], color='b', label='Z')
plt.title('Accelerometer Linear')
plt.legend(loc='upper right')
plt.show();
```

```python
plt.figure(figsize=(12, 6))
plt.plot(barometer['Time (s)'], barometer['X (hPa)'], color='r', label='X')
plt.title('Barometer')
plt.legend(loc='upper right')
plt.show();
```

```python
plt.figure(figsize=(12, 6))
plt.plot(gyroscope['Time (s)'], gyroscope['X (rad/s)'], color='r', label='X')
plt.plot(gyroscope['Time (s)'], gyroscope['Y (rad/s)'], color='g', label='Y')
plt.plot(gyroscope['Time (s)'], gyroscope['Z (rad/s)'], color='b', label='Z')
plt.title('Gyroscope')
plt.legend(loc='upper right')
plt.show();
```

```python
plt.figure(figsize=(12, 6))
plt.plot(magnetometer['Time (s)'], magnetometer['X (T)'], color='r', label='X')
plt.plot(magnetometer['Time (s)'], magnetometer['Y (T)'], color='g', label='Y')
plt.plot(magnetometer['Time (s)'], magnetometer['Z (T)'], color='b', label='Z')
plt.title('Magnetometer')
plt.legend(loc='upper right')
plt.ylim([-200, 200])
plt.show();
```
