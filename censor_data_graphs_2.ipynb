```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
SUBJECT = 'kaan' # ['yigit', 'onder', 'julian']
T_START = 21
DATA_PATH = f".\\data\\working_data\\{SUBJECT}\\"
INTERVALS = [0, 29, 35, 61.5, 67.5, 97.5, 103, 138, 147, 187.5, 191, 226.5, 232, 
             264.5, 269, 306, 314, 343, 348, 397.5, 403.5, 455, 462, 510]

accelerometer_linear = pd.read_csv(f"{DATA_PATH}Linear Accelerometer.csv")
gyroscope = pd.read_csv(f"{DATA_PATH}Gyroscope.csv")
magnetometer = pd.read_csv(f"{DATA_PATH}Magnetometer.csv")

print(accelerometer_linear.shape)
print(gyroscope.shape)
print(magnetometer.shape)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(accelerometer_linear['Time (s)'], accelerometer_linear['X (m/s^2)'], color='r', label='X')
plt.plot(accelerometer_linear['Time (s)'], accelerometer_linear['Y (m/s^2)'], color='g', label='Y')
plt.plot(accelerometer_linear['Time (s)'], accelerometer_linear['Z (m/s^2)'], color='b', label='Z')
[plt.axvline(x=_x+T_START, ls='--', lw=1.5, c='black') for _x in INTERVALS]
[plt.text(_x, 38, f"{_y+1}", fontsize='x-large', color='orange') for _x,_y in zip(INTERVALS[1::2], range(12))]
plt.title(f"{SUBJECT} - Accelerometer Linear")
plt.xlabel('time (s)')
plt.ylabel('(m/s^2)')
plt.legend(loc='upper right')
# plt.xlim([300, 600])
# plt.savefig(f"{SUBJECT}_accelerometer_linear.png", bbox_inches='tight')
plt.show();
```

```python
plt.figure(figsize=(15, 6))
plt.plot(gyroscope['Time (s)'], gyroscope['X (rad/s)'], color='r', label='X')
plt.plot(gyroscope['Time (s)'], gyroscope['Y (rad/s)'], color='g', label='Y')
plt.plot(gyroscope['Time (s)'], gyroscope['Z (rad/s)'], color='b', label='Z')
[plt.axvline(x=_x+T_START, ls='--', lw=1.5, c='black') for _x in INTERVALS]
[plt.text(_x+T_START-15, 15, f"{_y+1}", fontsize='x-large', color='orange') for _x,_y in zip(INTERVALS[1::2], range(12))]
plt.title(f"{SUBJECT} - Gyroscope")
plt.xlabel('time (s)')
plt.ylabel('(rad/s)')
plt.legend(loc='upper right')
# plt.xlim([10, 30])
# plt.savefig(f"{SUBJECT}_gyroscope.png", bbox_inches='tight')
plt.show();
```

```python
plt.figure(figsize=(15, 6))
plt.plot(magnetometer['Time (s)'], magnetometer['X (T)'], color='r', label='X')
plt.plot(magnetometer['Time (s)'], magnetometer['Y (T)'], color='g', label='Y')
plt.plot(magnetometer['Time (s)'], magnetometer['Z (T)'], color='b', label='Z')
[plt.axvline(x=_x+T_START, ls='--', lw=1.5, c='black') for _x in INTERVALS]
[plt.text(_x+T_START-15, 48, f"{_y+1}", fontsize='x-large', color='orange') for _x,_y in zip(INTERVALS[1::2], range(12))]
plt.title(f"{SUBJECT} - Magnetometer")
plt.xlabel('time (s)')
plt.ylabel('(T)')
plt.legend(loc='upper right')
# plt.xlim([10, 20])
# plt.savefig(f"{SUBJECT}_magnetometer.png", bbox_inches='tight')
plt.show();
```
